# IV. Preprocessing

```python
#Usuwanie niepotrzebnych kolum i uzupełnianie wartości nullowych

dane = dane.drop(['Id','4','5','7'], axis = 1)
dane.info()
```

```python
dane['3'] = dane['3'].fillna(dane['3'].mode()[0])
dane['6'] = dane['6'].fillna(dane['6'].mode()[0])
dane = dane.dropna()
dane.info()
```

```python
dane.head()
```

```python
X = dane.drop('target', axis = 1)
y = dane['target']
```

```python
#skalowanie cech
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)

X_scaled = scaler.transform(X)

X_scaled = pd.DataFrame(X_scaled, copy = True)
```

```python
#podział na zbiór treningowy i testowy
from sklearn.model_selection import train_test_split

X_train, X_test , y_train, y_test = train_test_split(X_scaled, y)
```

# V. Wyuczenie i dostrojenie modelu knn-R

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

regressor = KNeighborsRegressor(n_neighbors = 3)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

mean_squared_error(y_test, y_pred)
```

```python
regressor = KNeighborsRegressor(n_neighbors = 5)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

mean_squared_error(y_test, y_pred)
```

```python
regressor = KNeighborsRegressor(n_neighbors = 10, p=2)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

mean_squared_error(y_test, y_pred)
```

```python
regressor = KNeighborsRegressor(n_neighbors = 10, p=1)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

mean_squared_error(y_test, y_pred)
```

```python
#zwycięzca knn z k=7 i metryką Euklidesową
regressorkNN = KNeighborsRegressor(n_neighbors = 5, p=2)
regressorkNN.fit(X_train, y_train)
y_pred_kNN = regressorkNN.predict(X_test)

mean_squared_error(y_test, y_pred_kNN)
```

# VI. Wyuczenie i dostrojenie modelu regresji ze spadkiem gradientu

```python
from sklearn.linear_model import SGDRegressor

regressor = SGDRegressor(loss='squared_loss')
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

mean_squared_error(y_test, y_pred)
```

```python
regressor = SGDRegressor(max_iter=10, penalty=None)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

mean_squared_error(y_test, y_pred)
```

```python
regressor = SGDRegressor(max_iter=5, penalty=None)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

mean_squared_error(y_test, y_pred)
```

```python
regressor = SGDRegressor(max_iter=100, penalty=None)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

mean_squared_error(y_test, y_pred)
```

```python
regressor = SGDRegressor(eta0=0.1)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

mean_squared_error(y_test, y_pred)
```

```python
regressor = SGDRegressor(eta0=0.15)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

mean_squared_error(y_test, y_pred)
```

```python
regressor = SGDRegressor(max_iter=120, eta0=0.1)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

mean_squared_error(y_test, y_pred)
```

```python
#zwycięzca

GDregressor = SGDRegressor(max_iter=120, eta0=0.1)
GDregressor.fit(X_train, y_train)
y_pred_GDR = GDregressor.predict(X_test)

mean_squared_error(y_test, y_pred_GDR)
```

# VIII. Porównanie modeli i wnioski

Z modeli, które wziąłem pod uwagę najmniejszy błąd średniokwadratowy dla zbioru testowego uzyskał model regresji ze stochastycznym spadkiem gradientu z parametrami maxIter =100 i penalty=None gdzie tenże błąd wyniósł ok. 1.49.  W przypadku modelu kNN "udało się" zminimalizować błąd do ok. 1.76. Jednak przy predykcji widać różnice między modelami kNN i regresji SGD. Przede wszystkim model kNN generuje niższą wariancję od modelu regresji, odpowiednio ok. 1.29 i 1.98. Może być to spowodowane wrażliwością na outliery w regresji. Porównując te wyniki z wariancją dla najbliższej stacji do stacji docelowej(3.63) można jednak wysnuć wniosek, że oba modele są lekko "niedouczone" i możliwe, że wskazane byłoby dalsze "dostrajanie" ich lub użycie innych bardziej zaawansowanych modeli.
